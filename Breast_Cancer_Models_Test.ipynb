```python
data = [17.99,10.38,122.8,1001,0.1184,
        0.2776,0.3001,0.1471,0.2419,
        0.07871,1.095,0.9053,
        8.589,153.4,0.006399,0.04904,
        0.05373,0.01587,0.03003,
        0.006193,25.38,17.33,184.6,
        2019,0.1622,
        0.6656,0.7119,0.2654,0.4601,0.1189]
        
data = np.array(data)
data = tf.expand_dims(data,0)
#print(data.shape)

loaded_classifier = tf.keras.models.load_model('breast_cancer_diagnosis_features_model.h5')    
    
X_new = data 
y_pred = loaded_classifier.predict(X_new)

print('The probability of having a Maligant ' + str(y_pred[0][0]))
print('The probability of having a Bening ' + str(y_pred[0][1]))

# The 0 index is for malignant and 1 index is for Bening
if y_pred[0][0] > y_pred[0][1]:
    print('The case is Malignant')
    
if y_pred[0][0] < y_pred[0][1]:
    print('The case is Bening')
```
